### What is most likely outcome if you roll N dice symultaniously

```python
# Configure Jupyter so figures appear in the notebook
%matplotlib inline

# import library
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fucntion to simulate die roll
def die_roll(n):
    tot_sum = 0 
    for i in range(n):
        tot_sum += random.randint(1,6)
    return(tot_sum)

def die_sim(die_num, sim_cnt):
    count=0
    die_score=[]
    for i in range(sim_cnt):
        total=die_roll(die_num)
        die_score.append(total)
        
  
    plt.hist(die_score)
    plt.xlabel('Die Outcome')
    plt.ylabel('No of observationsCount')
    plt.title('Result of 1000 simulation')
        
```

```python
# When we roll 1 die the probability of getting each number is almost equal

die_sim(1, 1000)
```

```python
# When we roll 2 dies probability of getting 7 is highest
die_sim(2, 1000)
```

```python
# When we roll 5 dies probability of getting 15 is highest
die_sim(5, 1000)
```
